### 6) Realice un histograma con la población de los Municipios para cada una de las provincias.

```python
provincias = df['Provincia'].unique()

# Determinar el número de filas y columnas para organizar los subplots
num_provincias = len(provincias)
num_subplots_por_fila = 4
num_filas = (num_provincias - 1) // num_subplots_por_fila + 1
num_columnas = min(num_provincias, num_subplots_por_fila)

# Crear subplots
fig, axs = plt.subplots(num_filas, num_columnas, figsize=(15, 5*num_filas))

# Iterar sobre las provincias y crear un histograma para cada una
for i, provincia in enumerate(provincias):
    ax = axs[i // num_subplots_por_fila, i % num_subplots_por_fila]
    poblacion_provincia = df[df['Provincia'] == provincia]['Poblacion']
    ax.hist(poblacion_provincia, bins=10, color='blue', edgecolor='black')
    ax.set_title(provincia)
    ax.set_xlabel('Poblacion')
    ax.set_ylabel('Frecuencia')

plt.show()
```

### 8) Determine los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero

```python
sin_madrid = df[df['Municipio'] != 'Madrid'].copy()
madrid = df[df['Municipio'] == 'Madrid']
madridx = madrid['X'].values[0]
madridy = madrid['Y'].values[0]

sin_madrid.loc[:, 'Distancia'] = np.sqrt((madridx - sin_madrid['X'])**2 + (madridy - sin_madrid['Y'])**2)
```

```python
print("El Municipio más cercano a Madrid es:",sin_madrid[sin_madrid['Distancia'] == sin_madrid['Distancia'].min()]['Municipio'].values[0])
print("El Municipio más lejano a Madrid es:",sin_madrid[sin_madrid['Distancia'] == sin_madrid['Distancia'].max()]['Municipio'].values[0])
```
